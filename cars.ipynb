```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
new = pd.read_csv('new-used-cars-dataset.csv')
new
```

```python
new = new.drop(new[new.Mileages == "Mileage"].index)
new
```

```python
#new = df[df['used/certified'] != 'Used']
#new
```

```python
new[['Year','Name']] = new['Car Names'].str.split(n=1,expand=True)
new = new.drop('Car Names',axis=1)
new.Year = new.Year.astype(int)
new
```

```python
new[['Mileages','del']] = new['Mileages'].str.split(n=1,expand=True)
new[['Mileages1','Mileages2']] = new['Mileages'].str.split(",",n=1,expand=True)
new['Mileages'] = new['Mileages1'] + new['Mileages2']
new['Mileages'] = new['Mileages'].fillna(new.Mileages1)
new = new.drop(['Mileages1','Mileages2','del'],axis=1)
new.Mileages = new.Mileages.astype(int)
new
```

```python
new[['Reviews','del']] = new['Reviews'].str.split(n=1,expand=True)
new[['del','Reviews']] = new['Reviews'].str.split('(',n=1,expand=True)
new[['rev1','rev2']] = new['Reviews'].str.split(',',n=1,expand=True)
new['Reviews'] = new['rev1'] + new['rev2']
new['Reviews'] = new['Reviews'].fillna(new.rev1)

new = new.drop(['del','rev1','rev2'],axis=1)
new.Reviews = new.Reviews.astype(int)
new
```

```python
new[['Make','Model']] = new['Name'].str.split(n=1,expand=True)
new
```

```python
new[['price drop','del']] = new['price drop'].str.split(n=1,expand=True)
new[['del2','price drop']] = new['price drop'].str.split('$',n=1,expand=True)
new[['col1','col2']] = new['price drop'].str.split(',',n=1,expand=True)
new['price drop'] = new.col1 + new.col2
new['price drop'] = new['price drop'].fillna(new.col1)

new = new.drop(['del','del2','col1','col2'],axis=1)
new['price drop'] = new['price drop'].astype(int)
new
```

```python
new[['del','Price']] = new['Price'].str.split('$',n=1,expand=True)
new[['Price1','Price2']] = new['Price'].str.split(',',n=1,expand=True)
new['Price'] = new['Price1'] + new['Price2']
new.Price = new.Price.fillna(new.Price1)
new = new.dropna()
new.Price = new.Price.astype(int)

new = new.drop(['del','Price1','Price2'],axis=1)
new
```

```python
new['used/certified'].unique()
```

```python
new['used/certified'] = new['used/certified'].replace('Used','Replace Used')
new['used/certified'] = new['used/certified'].replace('Aston Martin Certified','Replace Certified')
new['used/certified'] = new['used/certified'].replace('Alfa Romeo Certified','Replace Certified')
new[['col1','used/certified']] = new['used/certified'].str.split(n=1,expand=True)
new['used_or_certified'] = new['used/certified']
new['used_or_certified'] = new['used_or_certified'].replace('Used','0')
new['used_or_certified'] = new['used_or_certified'].replace('Certified','1')
new['used_or_certified'] = new['used_or_certified'].astype(int)
new
```

```python
clean_df = new[['Make','Model','Year','used/certified','Price','price drop','Mileages','Ratings','Reviews']]
clean_df
```
